```python
#it is a 4sample so we are going to do anova test to find the pvalue
#if pvalue is greater than alpha value, it is fail to reject the null hypothesis
#if pvalue is less than or equal to alpha value, it is rejected the null hypothesis
#alpha = 5% = 0.05
```

```python
fstats, pvalue = stats.f_oneway(df.iloc[:,0],df.iloc[:,1],df.iloc[:,2],df.iloc[:,3])
```

```python
pvalue
```

```python
if pvalue <= 0.05:
    print('rejected the null hypothesis')
else:
    print("fail to reject the null hypothesis")
```
